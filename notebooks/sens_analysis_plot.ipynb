```python
# 相関係数ごとに1行4列の折れ線プロットを作成
correlation_types = ["kendall"]
datasets = ["RELATE", "AudioCaps", "MusicCaps", "Clotho"]
x_values = [0, 1, 2]  # 3つのデータポイント
font_size = 36
line_width = 6
marker_size = 20

# スタイル設定
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (16, 4)

for corr_type in correlation_types:
    fig, axes = plt.subplots(1, 4, figsize=(18, 8))

    for idx, dataset in enumerate(datasets):
        ax = axes[idx]

        # CLAPScore_humanのデータ
        clap_human_values = CLAPScore_human[dataset][corr_type]
        ax.plot(
            x_values,
            [v * 100 for v in clap_human_values],
            marker="s",
            linewidth=line_width,
            label="CLAPS._Human",
            color="#A23B72",
            markersize=marker_size,
        )

        # CLAPScore_laionのデータ
        clap_laion_values = CLAPScore_laion[dataset][corr_type]
        ax.plot(
            x_values,
            [v * 100 for v in clap_laion_values],
            marker="^",
            linewidth=line_width,
            label="CLAPS._LAION",
            color="#F18F01",
            markersize=marker_size,
        )

        # CLAPScore_msのデータ
        clap_ms_values = CLAPScore_ms[dataset][corr_type]
        ax.plot(
            x_values,
            [v * 100 for v in clap_ms_values],
            marker="d",
            linewidth=line_width,
            label="CLAPS._MS",
            color="#C73E1D",
            markersize=marker_size,
        )
        
        # Proposed_dictのデータ
        proposed_values = Proposed_dict[dataset][corr_type]
        ax.plot(
            x_values,
            [v * 100 for v in proposed_values],
            marker="o",
            linewidth=line_width,
            label="Proposed",
            color="#2E86AB",
            markersize=marker_size,
        )

        ax.set_title(dataset, fontsize=font_size)
        ax.tick_params(axis="y", labelsize=font_size)
        if idx == 0:
            ax.set_ylabel(r"Kendall's $\tau$ [%]", fontsize=font_size +2)
            ax.set_xlabel("# Acoustic events", fontsize=font_size)
        # ax.set_xticks(["1", "2", "3+"])
        # ax.legend(fontsize=8)
        ax.grid(True, alpha=1, linewidth=1)
        ax.set_xticks(x_values)
        ax.set_xticklabels(["1", "2", "3+"], fontsize=font_size)
        if dataset == "Clotho":
            ax.set_ylim(0, 35)
        elif dataset == "MusicCaps":
            ax.set_ylim(-20, 40)
        elif dataset == "AudioCaps":
            ax.set_ylim(0, 40)
        elif dataset == "RELATE":
            ax.set_ylim(3, 25)
        # ax.yaxis.set_major_formatter(FormatStrFormatter("%.1f"))

    handles, labels = axes[0].get_legend_handles_labels()
    fig.legend(
        handles,
        labels,
        loc="upper center",
        ncol=4,
        fontsize=font_size-2,
        frameon=True,  
        fancybox=True,
        bbox_to_anchor=(0.5, 1.15),
        borderpad=0.2,              # ★ 枠と中身の余白
        labelspacing=0.2,           # ★ 行間
        handletextpad=0.6,          # ★ 線と文字の間
        columnspacing=0.4,          # ★ 列間
    )
    
    plt.tight_layout()
    plt.show()
```
